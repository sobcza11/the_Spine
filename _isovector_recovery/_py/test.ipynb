```python
import boto3

BUCKET = "thespine-us-hub"

s3 = boto3.client(
    "s3",
    endpoint_url=os.getenv("R2_ENDPOINT"),
    aws_access_key_id=os.getenv("R2_ACCESS_KEY_ID"),
    aws_secret_access_key=os.getenv("R2_SECRET_ACCESS_KEY"),
    region_name=os.getenv("R2_REGION", "auto"),
)

keywords = ["cpi", "inflation", "pce", "breakeven", "fred"]

matches = []
paginator = s3.get_paginator("list_objects_v2")

for page in paginator.paginate(Bucket=BUCKET):
    for obj in page.get("Contents", []):
        key = obj["Key"]
        key_l = key.lower()
        if any(k in key_l for k in keywords):
            matches.append(key)

print(f"matches found: {len(matches)}")
for key in matches:
    print(key)
```

```python
import pandas as pd
from pathlib import Path

base = Path(r"C:\Users\Rand Sobczak Jr\_rts\3_AI\IsoVector\_py\r2_tmp")

files = {
    "us_fx_10y_spreads": base / "us_fx_10y_spreads.parquet",
    "us_ir_diff_canonical": base / "us_ir_diff_canonical.parquet",
    "us_equity_index_t1": base / "us_equity_index_t1.parquet",
    "us_sector_etf_t1": base / "us_sector_etf_t1.parquet",
    "us_ism_manu_pmi_by_industry_canonical": base / "us_ism_manu_pmi_by_industry_canonical.parquet",
    "us_ism_nonmanu_pmi_by_industry_canonical": base / "us_ism_nonmanu_pmi_by_industry_canonical.parquet",
}

def add_date_column(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    if "date" in df.columns:
        df["date_std"] = pd.to_datetime(df["date"], errors="coerce")
    elif "as_of_date" in df.columns:
        df["date_std"] = pd.to_datetime(df["as_of_date"], errors="coerce")
    elif "Date" in df.columns:
        df["date_std"] = pd.to_datetime(df["Date"], errors="coerce")
    else:
        df["date_std"] = pd.NaT
    return df

for name, path in files.items():
    print("\n" + "=" * 100)
    print(name.upper())
    print("=" * 100)

    df = pd.read_parquet(path)
    df = add_date_column(df)

    print("\nCOLUMNS:")
    print(df.columns.tolist())

    print("\nSHAPE:")
    print(df.shape)

    if "date_std" in df.columns and df["date_std"].notna().any():
        print("\nDATE RANGE:")
        print(df["date_std"].min(), "->", df["date_std"].max())

    likely_id_cols = [c for c in ["pair", "symbol", "base_ccy", "quote_ccy", "leaf_name", "industry", "sector"] if c in df.columns]
    if likely_id_cols:
        print("\nID COLUMNS:")
        for c in likely_id_cols:
            vals = df[c].dropna().astype(str).unique()[:10]
            print(f"{c}: {list(vals)}")

    print("\nTAIL:")
    print(df.tail())
```
